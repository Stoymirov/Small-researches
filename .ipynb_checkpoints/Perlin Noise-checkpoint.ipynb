## Key concepts ##

### Gradient ###

In computer graphics, a gradient is a gradual transition between colors or values. It's used to create smooth changes in shading, color, or intensity — which adds depth, realism, and style to images, shapes, or backgrounds.

![Image if gradient](https://th.bing.com/th/id/OIP.FPBUTR98JGfnUWARPpGzPQHaEK?rs=1&pid=ImgDetMain)

### Pseudo-random ###

Pseudo-random means something that appears random but is actually generated by a deterministic process — usually an algorithm. It's not truly random, but it’s random enough for most purposes like games, simulations, or graphics.

### Noise ###

​In computer graphics, noise refers to a function or algorithm that generates pseudo-random values. This means that at close points the values are coherent producing a smooth variation. Let us show a 1D variation of noise

```python
%matplotlib
import numpy as np
import matplotlib.pyplot as plp


def generate_1d_noice(length,step):
      x = np.arange(0, length, step)
      noise = np.random.rand(len(x))
      return x,noise

length = 10
step = 0.1

x,noise = generate_1d_noice(length,step)
plt.figure(figsize=(10,4))
plt.plot(x,noise,label = "1D Noise")
plt.title = "1D Noise"
plt.xlabel = "x"
plt.ylabel = "Noise Value"
plt.grid(True)
plt.legend()
plt.show()
```

Now I am going to try and do the same thing but make it smoother. In order to both understand a library and the method I am first going to show how it can be done using a library and then explain the maths behind it.

```python
from perlin_noise import PerlinNoise

# Initialize Perlin noise generator
noise = PerlinNoise(octaves=1, seed=42)

# Generate 1D Perlin noise
x = np.linspace(0, 10, 1000)
y = np.array([noise(val) for val in x])

# Plot the noise
plt.figure(figsize=(10, 4))
plt.plot(x, y, label="1D Perlin Noise")
plt.xlabel('x')
plt.ylabel("y")
plt.grid(True)
plt.legend()
plt.show()
```

We can easily observe that the graph is much smoother which is due to the fact that perlin noise is **pseudo-rando**.

Speaking of noise it would be a good idea to take a look at noise and see what kinds of noise there is.

## Types of noise ##



### White noise ###



White noise consists of random, uncorrelated values at each point, resulting in a highly irregular and grainy texture. It is used when absolute randomness is required.

### Gradient noise ###

 Gradient noise introduces spatial coherence by assigning gradient vectors to lattice points and interpolating between them. This results in smoother transitions and more natural patterns. As said it is ideal for generating terrain, clouds, and other natural phenomena.

## Fractals ##

This balance between randomness and order is beautifully exemplified in fractals, which are intricate patterns that repeat at every scale. Fractals showcase self-similarity, where smaller parts of a pattern resemble the whole, revealing a hidden structure within the apparent chaos. ## How does Perlin noise work ##

To begin with we need a grid. Having this grid we take a point that is randomly distributed in each cell of the grid, associated with a scalar value.

![Cell](https://iq.opengenus.org/content/images/2022/11/Scalar-point-Z.PNG)

Now for each point we are going to pseudo-randomly create 4 vectors coming from the corners of the cell.

![grid](https://th.bing.com/th/id/OIP.SW1kG6Yz4amsh9kf9NUwUAAAAA?rs=1&pid=ImgDetMain)

After that we need to calculate the distance vectors from the corners of the grid.

![grid distance vectors](https://th.bing.com/th/id/OIP.uePPvHxwWlceUj5DtOayNgHaD7?rs=1&pid=ImgDetMain)

After having all this we will simply have to apply a smooth function to make everything smoother. 

```python

# create a Perlin texture in 2D

import numpy as np
import matplotlib.pyplot as plot

def perlin(x, y, seed=0):
    # create a permutation table based on number of pixels
    # seed is the initial value we want to start with
    # we also use seed function to get same set of numbers
    # this helps to keep our perlin graph smooth
    np.random.seed(seed)
    ptable = np.arange(256, dtype=int)

    # shuffle our numbers in the table
    np.random.shuffle(ptable)

    # create a 2d array and then turn it one dimensional
    # so that we can apply our dot product interpolations easily
    ptable = np.stack([ptable, ptable]).flatten()
    
    # grid coordinates
    xi, yi = x.astype(int), y.astype(int)
   
    # distance vector coordinates
    xg, yg = x - xi, y - yi
    
    # apply fade function to distance coordinates
    xf, yf = fade(xg), fade(yg)
    
    # the gradient vector coordinates in the top left, top right, bottom left bottom right
   
    n00 = gradient(ptable[ptable[xi] + yi], xg, yg)
    n01 = gradient(ptable[ptable[xi] + yi + 1], xg, yg - 1)
    n11 = gradient(ptable[ptable[xi + 1] + yi + 1], xg - 1, yg - 1)
    n10 = gradient(ptable[ptable[xi + 1] + yi], xg - 1, yg)
    
    # apply linear interpolation i.e dot product to calculate average
    x1 = lerp(n00, n10, xf)
    x2 = lerp(n01, n11, xf)  
    return lerp(x1, x2, yf)  

def lerp(a, b, x):
    "linear interpolation i.e dot product"
    return a + x * (b - a)

# smoothing function,
# the first derivative and second both are zero for this function

def fade(f):
    
    return 6 * f**5 - 15 * f**4 + 10 * f**3

# calculate the gradient vectors and dot product
def gradient(c, x, y):
   
    vectors = np.array([[0, 1], [0, -1], [1, 0], [-1, 0]])
    gradient_co = vectors[c % 4]
    return gradient_co[:, :, 0] * x + gradient_co[:, :, 1] * y

# create evenly spaced out numbers in a specified interval
lin_array = np.linspace(1, 10, 500, endpoint=False)

# create grid using linear 1d arrays
x, y = np.meshgrid(lin_array, lin_array)  

# generate graph
plot.imshow(perlin(x, y, seed=2), origin = 'upper')

plot.show()

```
